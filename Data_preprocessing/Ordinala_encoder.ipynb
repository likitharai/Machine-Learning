```python
import pandas as pd
```

```python
df=pd.DataFrame({'size':['xs','xl','xs','s','m','l','xl','s','xs','m','xl','xl']})
```

```python

df
```

```python
ord_data=[['xs','s','m','l','xl']]
```

## sklearn library

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
OE=OrdinalEncoder(categories=ord_data)
OE.fit(df[['size']])
```

```python
df['size_en']=OE.transform(df[['size']])
df
```

## map function

```python
ord_data1={'xs':1,'s':2,'m':3,'l':4,'xl':5}
```

```python
df['size_en1']=df['size'].map(ord_data1)
```
