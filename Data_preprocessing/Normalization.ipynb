```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('loan.csv')
data.head()
```

```python
data.describe()
```

```python
from sklearn.preprocessing import MinMaxScaler
ms=MinMaxScaler()
ms.fit(data[['CoapplicantIncome']])
```

```python
data['CoapplicantIncome_ms']=pd.DataFrame(ms.transform(data[['CoapplicantIncome']]),columns=['x'])
```

```python
data.head()
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.title('before')
sns.distplot(data['CoapplicantIncome'])
plt.subplot(1,2,2)
plt.title('after')

sns.distplot(data['CoapplicantIncome_ms'])
plt.show()
```
